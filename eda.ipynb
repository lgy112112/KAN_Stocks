```python
import plotly.graph_objects as go
import pandas as pd

# 读取 CSV 文件
file_path = '600519.SH.CSV'
stock_data = pd.read_csv(file_path, encoding='gbk')

# 转换日期列为 datetime 格式
stock_data['日期'] = pd.to_datetime(stock_data['日期'])

# 创建图表
fig = go.Figure()

fig.add_trace(go.Scatter(x=stock_data['日期'], y=stock_data['收盘价(元)'],
                         mode='lines',
                         name='收盘价'))

# 设置图表标题和标签
fig.update_layout(
    title='贵州茅台 (600519.SH) 收盘价走势图',
    xaxis_title='日期',
    yaxis_title='收盘价 (元)',
    font=dict(family="SimHei", size=18)
)

# 显示图表
fig.show()
```

```python
# Check for NaN values in each column
nan_summary = stock_data.isna().sum()


nan_summary
```

```python
# Drop the 'Unnamed: 18' column
stock_data_cleaned = stock_data.drop(columns=['Unnamed: 18'])

# Check if the rows with NaN values are the same across the mentioned columns
nan_rows = stock_data_cleaned[stock_data_cleaned.isna().any(axis=1)]

# Verify if all rows with NaN values are the same
unique_nan_rows = nan_rows.dropna().empty

nan_rows
# unique_nan_rows
```

```python
import plotly.express as px
import plotly.graph_objects as go

# 1. 收盘价趋势图
fig1 = px.line(stock_data_cleaned, x='日期', y='收盘价(元)', title='贵州茅台 (600519.SH) 收盘价趋势图')
fig1.update_layout(xaxis_title='日期', yaxis_title='收盘价 (元)')

# 2. 交易量趋势图
fig2 = px.line(stock_data_cleaned, x='日期', y='成交量(股)', title='贵州茅台 (600519.SH) 交易量趋势图')
fig2.update_layout(xaxis_title='日期', yaxis_title='成交量 (股)')

# 4. 涨跌幅分布图
fig3 = px.histogram(stock_data_cleaned, x='涨跌幅(%)', nbins=50, title='涨跌幅(%) 分布图')
fig3.update_layout(xaxis_title='涨跌幅(%)', yaxis_title='频率')

# 5. 年度趋势
annual_trends = stock_data_cleaned.groupby(stock_data_cleaned['日期'].dt.year).agg({'收盘价(元)': 'mean', '成交量(股)': 'sum'}).reset_index()
fig4 = go.Figure()
fig4.add_trace(go.Scatter(x=annual_trends['日期'], y=annual_trends['收盘价(元)'], mode='lines+markers', name='收盘价均值'))
fig4.update_layout(title='年度收盘价均值趋势图', xaxis_title='年份', yaxis_title='收盘价均值 (元)')

fig5 = go.Figure()
fig5.add_trace(go.Scatter(x=annual_trends['日期'], y=annual_trends['成交量(股)'], mode='lines+markers', name='成交量总和'))
fig5.update_layout(title='年度成交量总和趋势图', xaxis_title='年份', yaxis_title='成交量总和 (股)')

# 6. 月度趋势
monthly_trends = stock_data_cleaned.groupby(stock_data_cleaned['日期'].dt.to_period('M')).agg({'收盘价(元)': 'mean', '成交量(股)': 'sum'}).reset_index()
fig6 = go.Figure()
fig6.add_trace(go.Scatter(x=monthly_trends['日期'].astype(str), y=monthly_trends['收盘价(元)'], mode='lines+markers', name='收盘价均值'))
fig6.update_layout(title='月度收盘价均值趋势图', xaxis_title='月份', yaxis_title='收盘价均值 (元)')
fig6.update_xaxes(tickangle=45)

fig7 = go.Figure()
fig7.add_trace(go.Scatter(x=monthly_trends['日期'].astype(str), y=monthly_trends['成交量(股)'], mode='lines+markers', name='成交量总和'))
fig7.update_layout(title='月度成交量总和趋势图', xaxis_title='月份', yaxis_title='成交量总和 (股)')
fig7.update_xaxes(tickangle=45)

# 7. 相关性热图
correlation_matrix = stock_data_cleaned[['开盘价(元)', '最高价(元)', '最低价(元)', '收盘价(元)', '成交量(股)']].corr()
fig8 = px.imshow(correlation_matrix, text_auto=True, aspect='auto', title='价格与成交量的相关性热图')

# 展示所有图表
fig1.show()
fig2.show()
fig3.show()
fig4.show()
fig5.show()
fig6.show()
fig7.show()
fig8.show()
```

```python
stock_data_cleaned['year'] = stock_data_cleaned['日期'].dt.year
stock_data_cleaned['month'] = stock_data_cleaned['日期'].dt.month
stock_data_cleaned['day'] = stock_data_cleaned['日期'].dt.day
stock_data_cleaned['day_of_week'] = stock_data_cleaned['日期'].dt.dayofweek

stock_data = stock_data_cleaned
stock_data
```

```python
# stock_data = stock_data_cleaned

stock_data['MA_5'] = stock_data['收盘价(元)'].rolling(window=5).mean()
stock_data['MA_10'] = stock_data['收盘价(元)'].rolling(window=10).mean()

stock_data
```

```python
stock_data['daily_return'] = stock_data['收盘价(元)'].pct_change()
stock_data['volatility_5'] = stock_data['daily_return'].rolling(window=5).std()
stock_data['volatility_10'] = stock_data['daily_return'].rolling(window=10).std()
```

```python
def calculate_rsi(data, window):
    delta = data.diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()
    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

stock_data['RSI_14'] = calculate_rsi(stock_data['收盘价(元)'], window=14)
```
